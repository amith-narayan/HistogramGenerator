```python
# inputs and outputs for the NN
t_columns = ['pocam_string', 'pocam_om', 'string', 'om', 'dist', 'zenith', 'abs', 'sca', 'domeff', 'p0', 'p1']
x_columns = df.columns[:30]
```

```python
x = df[x_columns].values
t = df[t_columns].values
x_train, x_test, t_train, t_test = train_test_split(x, t, test_size=0.2, random_state=42)
```

```python
activation='relu'

inputs = tf.keras.Input(shape=(t_train.shape[1],))
h = tf.keras.layers.Dense(64, activation=activation)(inputs)
h = tf.keras.layers.Dense(64, activation=activation)(h)
h = tf.keras.layers.Dense(256, activation=activation)(h)
h = tf.keras.layers.Dense(256, activation=activation)(h)
h = tf.keras.layers.Dense(512, activation=activation)(h)
h = tf.keras.layers.Dense(512, activation=activation)(h)
outputs = tf.keras.layers.Dense(x_train.shape[1], activation='softplus')(h)

model = tf.keras.Model(inputs=inputs, outputs=outputs)
```

```python
model.summary()
```

```python
batch = 3072

optimizer = tf.keras.optimizers.Adam(0.01)
model.compile(loss='poisson', optimizer=optimizer)
hist = model.fit(t_train, x_train, epochs=500, verbose=1, validation_data=(t_test, x_test), batch_size=batch)

optimizer = tf.keras.optimizers.Adam(0.0001)
model.compile(loss='poisson', optimizer=optimizer)
hist = model.fit(t_train, x_train, epochs=200, verbose=1, validation_data=(t_test, x_test), batch_size=batch)

optimizer = tf.keras.optimizers.Adam(0.000001)
model.compile(loss='poisson', optimizer=optimizer)
hist = model.fit(t_train, x_train, epochs=200, verbose=1, validation_data=(t_test, x_test), batch_size=batch)

optimizer = tf.keras.optimizers.Adam(0.00000001)
model.compile(loss='poisson', optimizer=optimizer)
hist = model.fit(t_train, x_train, epochs=100, verbose=1, validation_data=(t_test, x_test), batch_size=batch)
```

```python
# make a few random plots for a random datapoints
fig, ax = plt.subplots(4,4, figsize=(20,20))

ax = ax.flatten()

for j in range(16):

    axis = ax[j]
    rand_idx = np.random.randint(0, t_test.shape[0])
    inp = t_test[rand_idx]
    outp = x_test[rand_idx]
    
    pred_inp = inp[np.newaxis,:]
    pred = model.predict(pred_inp)[0]

    axis.plot(outp, label='Data')
    axis.plot(pred, label='NN expectation')
ax[-1].legend()
```
